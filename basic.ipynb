## PYPLOT
- Draw a line in a diagram from position (0,0) to position (6,250)

````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([0,6])#create a array x point (0->6)
ypoints = np.array([0,250])#create a array y point (0->250)
plt.plot(xpoints,ypoints)#Draw plot
plt.show()#Show
````

```plot()``` function is used to draw points in a diagram.

By default, the ```plot()``` draw a line from point to point.

Parameter 1 is a array containing the point on the x-axis.

Parameter 2 is a array containing the point on the y-axis.

## Marker
We can use keyword ```maker``` to emphasize each pointer with a specify maker

````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([0,6,8,1])#create a array x point (0->6)
ypoints = np.array([0,250,150,100])#create a array y point (0->250)
plt.plot(xpoints,ypoints,marker  = 'o')#use marker maker = 'optional'('o','*',anything)
plt.show()#Show
````

## Format String fmt
This parameter called fmt, and written with syntax: ```marker|line|color```

Line Syntax	       Description
'-'               	Solid lin	
':               '	Dotted line	
'-           -'	Dashed line	
'           -.'	Dashed/dotted line

````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([0,6,8,1])#create a array x point (0->6)
ypoints = np.array([0,250,150,100])#create a array y point (0->250)
plt.plot(xpoints,ypoints,'o-.y')#marker|line|color
plt.show()#Show
````

## Matplotlib Labels and Title
Use ```xlabel``` and ```ylabel``` to set the label for x- and y-axis

Use ```title``` to set title for chart

````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([15,20,25,30,35,45])#create a array x point (0->6)
ypoints = np.array([5,100,150,100])#create a array y point (0->250)
plt.plot(xpoints,'o-.y',mec = 'r',ms = 20)#mec = marker edge color
plt.plot(ypoints,'o-.b',mec = 'black',ms = 20)#mec = marker edge color
plt.xlabel("Temperture")
plt.ylabel("Huminity")
plt.title("Reported Climate change 2024")
plt.show()#Show
````

## Set Font properties for titel and labels
Use ```fontdict``` to set font for title


````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([15,20,25,30,35,45])#create a array x point (0->6)
ypoints = np.array([5,100,150,100])#create a array y point (0->250)
font1= {'family':'serif','color':'blue','size':'20'}
font2= {'family':'serif','color':'darkred','size':'15'}
plt.plot(xpoints,'o-.y',mec = 'r',ms = 20)#mec = marker edge color
plt.plot(ypoints,'o-.b',mec = 'black',ms = 20)#mec = marker edge color
plt.xlabel("Temperture",fontdict=font2)
plt.ylabel("Huminity",fontdict=font2)
plt.title("Reported Climate change 2024",fontdict=font1)
plt.show()#Show
````

## Matplotlib Adding Grid Lines
Use ```grid()``` function to add grid line to lines to the plot

````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([15,20,25,30,35,45])#create a array x point (0->6)
ypoints = np.array([5,100,150,100,150,200])#create a array y point (0->250)
font1= {'family':'serif','color':'blue','size':'20'}
font2= {'family':'serif','color':'darkred','size':'15'}
plt.plot(xpoints,ypoints,'o-.y',mec = 'r',ms = 20)#mec = marker edge color
plt.grid()#grid 
plt.xlabel("Temperture",fontdict=font2)
plt.ylabel("Huminity",fontdict=font2)
plt.title("Reported Climate change 2024",fontdict=font1)
plt.show()#Show
````

## Set Line Properties for the Grid
Use ```grid(color = 'color', linestyle = 'linestyle', linewidth = number)``` to set line 

````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([15,20,25,30,35,45])#create a array x point (0->6)
ypoints = np.array([5,100,150,100,150,200])#create a array y point (0->250)
font1= {'family':'serif','color':'blue','size':'20'}
font2= {'family':'serif','color':'darkred','size':'15'}
plt.plot(xpoints,ypoints,'o-.y',mec = 'r',ms = 20)#mec = marker edge color
plt.grid(c = 'green', ls = '--', lw = 0.5)
plt.xlabel("Temperture",fontdict=font2)
plt.ylabel("Huminity",fontdict=font2)
plt.title("Reported Climate change 2024",fontdict=font1)
plt.show()#Show
````

## Matplotlib Subplot
Use ```subplot``` function to draw multiple plot

```plt.plot(1,2,1)``` has 1 row, 2 colums, and this plot is the first plot

```plt.plot(1,2,2)``` has 1 row, 2 colums, and this plot is the second plot

````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = ([20,15,25,30,35,38])
ypoints =([50,60,55,35,70,80])
font1={'family':'serif','color':'blue','size':'15'}
font2= {'family':'serif','color':'darkred','size':'10'}
plt.subplot(1,2,1)
plt.plot(xpoints,ypoints,'o-.y',mec='r',ms=5)
plt.xlabel("Temperture",fontdict=font1)
plt.ylabel("Huminity",fontdict=font1)
plt.title("Reported Climate change 2024",loc='center',fontdict=font2)

xpoint = ([7,5,4,8,9,2])
ypoint =([50,55,34,60,70,75])
plt.subplot(1,2,2)
plt.plot(xpoint,ypoint,'o-.b',mec='r',ms=5)
plt.title("Reported Climate change 2025",loc='center',fontdict=font2)
plt.suptitle("REPORTED")
plt.show()
````

## Alpha
You can adjust the transparency of the dots with the alpha argument.

````python
import matplotlib.pyplot as plt
import numpy as np
xpoints = ([20,15,25,30,35,38])
ypoints =([50,60,55,35,70,80])
font1={'family':'serif','color':'blue','size':'15'}
font2= {'family':'serif','color':'darkred','size':'10'}
plt.xlabel("Temperture",fontdict=font1)
plt.ylabel("Huminity",fontdict=font1)

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='Accent',s=100,alpha=0.2)
plt.colorbar()

plt.show()
````

## Matplotlib Bars

````python
import matplotlib.pyplot as plt
import numpy as np
x= np.array(['2010','2015','2020','2025'])
y=np.array([86.9,91.7,97.6,104])
font1={'family':'serif','color':'darkred','size':'20'}
font2={'family':'serif','color':'blue','size':'15'}
plt.xlabel("Year",fontdict=font2)
plt.ylabel("Milion People",fontdict=font2)
plt.bar(x,y,color='gold',width=0.5)
plt.title("Population Report",fontdict=font1)
plt.show()
````
