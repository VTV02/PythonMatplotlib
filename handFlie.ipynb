## Create Graph from the Text File

````python
import matplotlib.pyplot as plt 
  
names = [] 
age = [] 
  
f = open('/kaggle/input/nameage-txt/namage.txt','r') 
for row in f: 
    row = row.split(' ') 
    names.append(row[0]) 
    age.append(int(row[1])) 
  
plt.bar(names, age, color = 'hotpink', label = 'Age') 
font1={'family':'serif','color':'blue','size':'15'}
font2= {'family':'serif','color':'darkred','size':'20'}
plt.xlabel('Name', fontdict=font1) 
plt.ylabel('Age', fontdict=font1) 
  
plt.title('Student Information Report',fontdict=font2) 
plt.legend() 
plt.show() 
````

````python
import matplotlib.pyplot as plt 

names = [] 
work = [] 
for line in open('/kaggle/input/percent-txt/namage.txt', 'r'): 
    Data = line.split() 
    names.append(Data[0]) 
    work.append(float(Data[1].replace('%', '')))  

colors = ['yellow', 'b', 'green', 'cyan', 'red'] 
	
# Sử dụng autopct để hiển thị chính xác phần trăm từ dữ liệu gốc
plt.pie(work, labels=names, colors=colors, startangle=90, 
        shadow=True, radius=1.2, 
        autopct=lambda p: f'{work[names.index(plt.gca().texts[-1].get_text())]}%') 
plt.show()
````

````python
from matplotlib import pyplot as plt
Days=[]
Temperature=[]
for line in open('/kaggle/input/weather-txt/weather.txt','r'):
    data=line.split()
    Days.append(data[0])
    Temperature.append(int(data[1]))
font1={'family':'serif','color':'blue','size':'15'}
font2={'family':'serif','color':'darkred','size':'20'}
plt.xlabel('Day of month',fontdict=font1)
plt.ylabel('Temperature Celsius Degree',fontdict=font1)
plt.title("Weather Reported 2024 in Asia",fontdict=font2)
plt.plot(Days,Temperature,'o-.',c='hotpink')
plt.grid(axis='y',c='g')
plt.show()
    
````
